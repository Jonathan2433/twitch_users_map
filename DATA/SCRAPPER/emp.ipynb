```python
import authenticator
import twitch_api_login
import call_api_service
```

```python
clientId = twitch_api_login.Client_ID
clientSecret = twitch_api_login.Secret

auth = authenticator.Auth(clientId, clientSecret)
access_token = auth.get_auth()

access_token
```

```python
ca = call_api_service.Call_api_service(clientId, access_token)

dev_optic_id = ca.get_user_id('dev_optic')

followers = ca.get_user_follows(str(dev_optic_id))

print(followers)
print(len(followers))
```
